```python
import pandas as pd

#Load the dataset
df = pd.read_csv("all_patient_features_db2.csv")  # Update filename if needed

#Fill missing values with mode (safely)
for column in df.columns:
    if df[column].isnull().any():
        mode_value = df[column].mode()
        if not mode_value.empty:
            df[column] = df[column].fillna(mode_value[0])  # safer assignment

#Save the cleaned dataset
df.to_csv("mode_imputed_patient_features.csv", index=False)

#Preview the result
print("Missing values filled with mode. Cleaned dataset saved.")
df.head()
```
